# Fusionar los tres Json

```python
import json
import pandas as pd
import numpy as np
import openpyxl 
```

```python
# Cargar los tres archivos JSON
with open('datos_relojes_chrono24.json', 'r', encoding='utf-8') as f1, \
     open('datos_relojes_sothebys.json', 'r', encoding='utf-8') as f2, \
     open('datos_relojes_farfetch.json', 'r', encoding='utf-8') as f3:
    data1 = json.load(f1)
    data2 = json.load(f2)
    data3 = json.load(f3)

# Fusionar las listas de los JSON
fusion = data1 + data2 + data3

# Guardar el resultado en un nuevo archivo JSON
with open('relojes.json', 'w', encoding='utf-8') as f_out:
    json.dump(fusion, f_out, ensure_ascii=False, indent=4)

print("Los archivos JSON han sido fusionados con éxito.")

```

# Limpieza y Normalizacion del DF

```python
ruta_json = "relojes.json"
df_relojes = pd.read_json(ruta_json)
df_relojes
```

```python
# Cargar el JSON fusionado
with open('relojes.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Crear un DataFrame a partir del JSON
df = pd.DataFrame(data)

# Columnas de interés
columnas_interes = [
    'Marca', 
    '(Reloj) Modelo', 
    "Modelo", 
    'Model', 
    "Precio",
    'Año', 
    'Year', 
    'Año de fabricación', 
    'Número de referencia', 
    'Reference Number', 
    'Estado', 
    'Condition Report', 
    'Contenido de la entrega', 
    'Accesorios incluidos', 
    'Box'
]

# Filtrar solo las columnas de interés que existan en el DataFrame
df_filtrado = df[[col for col in columnas_interes if col in df.columns]]

# Asegurarse de que todas las columnas de interés estén presentes en el DataFrame, rellenando las faltantes con NaN
for columna in columnas_interes:
    if columna not in df_filtrado.columns:
        df_filtrado[columna] = np.nan

df_filtrado
```

```python
df = df_filtrado.copy()

# Limpiar los nombres de las columnas para evitar problemas con espacios invisibles
df.columns = df.columns.str.strip()

# Unificar las columnas '(Reloj) Modelo', 'Modelo' y 'Model' en una sola columna llamada 'Modelo'
df['Modelo'] = df['(Reloj) Modelo'].fillna('') + ' ' + df['Modelo'].fillna('') + ' ' + df['Model'].fillna('')
df['Modelo'] = df['Modelo'].str.strip().replace('', np.nan)

# Eliminar las columnas originales después de la combinación
df = df.drop(columns=['(Reloj) Modelo', 'Model'], errors='ignore')

# Unificar las columnas 'Año', 'Year' y 'Año de fabricación' en una sola columna llamada 'Año'
df['Año'] = df['Año'].fillna('') + ' ' + df['Year'].fillna('') + ' ' + df['Año de fabricación'].fillna('')
df['Año'] = df['Año'].str.strip().replace('', np.nan)

# Eliminar las columnas originales después de la combinación
df = df.drop(columns=['Year', 'Año de fabricación'], errors='ignore')

# Unificar las columnas 'Número de referencia' y 'Reference Number' en una sola columna llamada 'Número de referencia'
df['Número de referencia'] = df['Número de referencia'].fillna('') + ' ' + df['Reference Number'].fillna('')
df['Número de referencia'] = df['Número de referencia'].str.strip().replace('', np.nan)

# Eliminar las columnas originales después de la combinación
df = df.drop(columns=['Reference Number'], errors='ignore')

# Unificar las columnas 'Estado' y 'Condition Report' en una sola columna llamada 'Estado'
df['Estado'] = df['Estado'].fillna('') + ' ' + df['Condition Report'].fillna('')
df['Estado'] = df['Estado'].str.strip().replace('', np.nan)

# Eliminar las columnas originales después de la combinación
df = df.drop(columns=['Condition Report'], errors='ignore')

# Unificar las columnas 'Contenido de la entrega', 'Accesorios incluidos', y 'Box' en una sola columna llamada 'Accesorios'
df['Accesorios'] = (
    df['Contenido de la entrega'].fillna('') + ' ' +
    df['Accesorios incluidos'].fillna('') + ' ' +
    df['Box'].fillna('')
)
df['Accesorios'] = df['Accesorios'].str.strip().replace('', np.nan)

# Eliminar las columnas originales después de la combinación
df = df.drop(columns=['Contenido de la entrega', 'Accesorios incluidos', 'Box'], errors='ignore')

# Reorganizar las columnas de acuerdo a tus necesidades
columnas_reordenadas = ['Marca', 'Modelo', 'Precio', 'Año', 'Número de referencia', 'Estado', 'Accesorios'] + [col for col in df.columns if col not in ['Marca', 'Modelo', 'Precio', 'Año', 'Número de referencia', 'Estado', 'Accesorios']]
df = df[columnas_reordenadas]

# Ver el DataFrame resultante
df
```

```python
# Guardar el DataFrame en un archivo JSON
df.to_json('relojes_definitivo.json', orient='records', lines=False, force_ascii=False)

# Guardar el DataFrame en un archivo Excel
df.to_excel('relojes_definitivo.xlsx', index=False)
```
